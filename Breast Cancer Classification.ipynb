Importing the Dependencies

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import GridSearchCV
```

```python
dataset = sklearn.datasets.load_breast_cancer()
```

```python

data_frame = pd.DataFrame(dataset.data, columns = dataset.feature_names)
```

```python
data_frame['label'] = dataset.target
```

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']

```

data spilting

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
def scores(y_test, y_pred, model):
    cr = classification_report(y_test, y_pred)
    print('Classification report:\n', cr, '\n')

    # Scores
    score_train = round(model.score(X_train, Y_train), 3)
    score_test = round(model.score(X_test, y_test), 3)
    print('Training Score: ', score_train)
    print('Testing Score: ', score_test)

    # Confusion Matrix
    conf_matrix = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(10, 7))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['TRUE', 'FALSE'], yticklabels=['TRUE', 'FALSE'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.show()
```

Model Training

Logistic_Regression MODEL

```python
logistic_regression_pipeline = Pipeline([
    
    ('Scaler',StandardScaler()),
    ('classifier', LogisticRegression())
])


logistic_regression_pipeline.fit(X_train, Y_train)


pred = logistic_regression_pipeline.predict(X_test)


scores(Y_test, pred, logistic_regression_pipeline)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_frame, x='mean radius')
plt.title('Boxplot of Mean Radius')
plt.show()
```

Random_Forest MODEL

```python
random_forest_pipeline = Pipeline([
     ('Scaler',StandardScaler()),
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=10, max_depth=10, n_jobs=-1,min_samples_split=30,min_samples_leaf=38,max_features='log2'))
])
random_forest_pipeline.fit(X_train, Y_train)
rfc_pred = random_forest_pipeline.predict(X_test)
scores(Y_test, rfc_pred, random_forest_pipeline)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_frame, x='mean radius')
plt.title('Boxplot of Mean Radius')
plt.show()
```

SVM MODEL

```python
svm_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', SVC(random_state=10))
])
param_grid = {
    'classifier__C': [0.1, 1, 10, 100],
    'classifier__gamma': [1, 0.1, 0.01, 0.001],
    'classifier__kernel': ['linear', 'rbf', 'poly', 'sigmoid']
}

grid_search = GridSearchCV(estimator=svm_pipeline, param_grid=param_grid,
                           cv=5, n_jobs=-1, scoring='accuracy')
grid_search.fit(X_train, Y_train)

best_params = grid_search.best_params_
print(f"Best parameters: {best_params}")

best_model = grid_search.best_estimator_
svm_pred = best_model.predict(X_test)
```

```python
scores(Y_test, svm_pred, best_model)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_frame, x='mean radius')
plt.title('Boxplot of Mean Radius')
plt.show()
```
